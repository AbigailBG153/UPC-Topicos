11. Print the names of all the forwards in the data file.

```python
!awk -F'\t' '$2 == "Forward" {print $1}' players.tsv
```

12. Calculate and print the total number of goals scored by all players.

```python
!awk -F'\t' '{sum += $3} END {print sum}' players.tsv
```

13. For each player, calculate their "shot accuracy" as the percentage of shots on goal that resulted in a goal (Goals / ShotsOnGoal * 100). Print the player's name and their shot accuracy. Be mindful of potential division by zero if a player has 0 shots on goal.

```python
!awk -F'\t' '{if ($5 != 0) {print $1, $3 / $5 * 100} else {print $1, "No shots on goal"}}' players.tsv
```

14. Find and print the name of the player who has the most assists. If there's a tie, you can print any one of them.

```python
!awk -F'\t' 'NR > 1 {if ($4 > max) {max = $4; name = $1}} END {print name}' players.tsv
```

15. List the names of all players who have played more than 27 games.

```python
!awk -F'\t' '$6 > 27 {print $1}' players.tsv
```

18. (5 points) Make a small brief of the paper explaining the main elements.

Este artículo presenta un enfoque híbrido que combina programación por restricciones (CP) y aprendizaje profundo (DL) para resolver el Problema de Programación Flexible de Taller (FJSSP). El objetivo es mejorar la eficiencia de las soluciones en tiempo real sin la necesidad de explorar exhaustivamente grandes espacios de acción, como en el aprendizaje por refuerzo profundo (DRL).

El artículo resalta que CP es muy eficaz para problemas pequeños, pero se vuelve computacionalmente intensivo para problemas más grandes. Para superar esto, proponen un modelo que utiliza CP para generar soluciones iniciales óptimas y luego entrena un modelo de aprendizaje profundo usando esas soluciones, mejorando así la eficiencia sin explorar excesivamente. Además, integran un predictor de capacidad de CP, que evalúa cuándo usar CP en tiempo real o cuándo recurrir a DL para simplificar el problema.

En 113 de 148 instancias probadas, el enfoque híbrido obtuvo los mejores resultados o empató con otros métodos, destacándose por su capacidad de resolver problemas más rápidamente y con mayor precisión. Además, el modelo híbrido fue más eficiente que enfoques meta-heurísticos como algoritmos genéticos y algoritmos Jaya.
